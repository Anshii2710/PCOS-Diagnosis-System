```python
import pandas as pd
```

```python
df=pd.read_csv('/content/pcos_dataset.csv')
df.head()
```

```python
df.sample(frac=1,random_state=42)
```

```python
# x_train = df.drop('PCOS_Diagnosis',axis=1)
train,test=train_test_split(df,test_size=0.2)
# y_train = df['PCOS_Diagnosis']
```

```python
x_train=train.drop('PCOS_Diagnosis',axis=1)
y_train=train['PCOS_Diagnosis']
```

```python
x_test=test.drop('PCOS_Diagnosis',axis=1)
y_test=test['PCOS_Diagnosis']
```

```python
model=LogisticRegression(C=0.1,max_iter=500)
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_predict)
accuracy
```
